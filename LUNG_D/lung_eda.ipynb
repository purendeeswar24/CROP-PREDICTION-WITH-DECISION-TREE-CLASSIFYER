```python
def out(col):
    q1 , q3 = np.percentile(col,[25,75])
    iqr=q3-q1
    ll=q1-1.5*iqr
    ul=q3+1.5*iqr
    return ll , ul
```

```python
a=["AGE"]
for i in a:
   ll , ul = out(df[i])
   df[i]  = np. where (df[i]<ll , ll, df[i])
   df[i] = np.where(df[i] >ul , ul , df[i])
   
```

```python
x=df.drop(columns=["LUNG_CANCER"])
y=df["LUNG_CANCER"]
```
